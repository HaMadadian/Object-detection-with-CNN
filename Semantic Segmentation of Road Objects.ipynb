## To start I want to check how create a colored image frm a grayscale image.

```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
import os

## Step 1: Mount Google Drive
#drive.mount('/content/drive')

# Step 2: Define the file path
# Assuming your Colab notebook and the image are in the same folder
notebook_path = "/content/drive/MyDrive/Object-detection-with-CNN/"
file_name = "Colorfull.jpeg"  # Change to your image file name
file_path = os.path.join(notebook_path, file_name)

# Step 3: Read and convert the image to grayscale
image = cv2.imread(file_path)  # Read the image from Google Drive
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

# Save or display the result
cv2.imwrite('Grayscale_image.jpeg', gray_image)

# Step 4: Display the original and grayscale images
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

# Display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.show()
```

```python
# Convert the grayscale image to a color image (BGR format)
color_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)

## Save or display the result
cv2.imwrite('color_image.jpeg', color_image)

# Step 4: Display the original and grayscale images
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

# Display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(color_image)
plt.title("Converted Color Image")
plt.axis('off')

plt.show()
```

```python
# Convert the grayscale image to a color image (BGR format)
color_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)

# Create a color image with a custom color, e.g., a bluish tint
color_image[:, :, 0] = gray_image // 2  # Blue channel
color_image[:, :, 1] = gray_image // 2  # Green channel (dimmed)
color_image[:, :, 2] = gray_image // 1  # Red channel (dimmed)


## Save or display the result
cv2.imwrite('color_image.jpeg', color_image)

# Step 4: Display the original and grayscale images
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

# Display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title("Converted Color Image")
plt.axis('off')

plt.show()
```
